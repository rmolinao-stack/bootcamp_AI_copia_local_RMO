# Proveedores de IA. Acceso a distintos modelos LLM y ejemplos

## Panorama: proveedores de IA en el mercado

Cuando construís un asistente con LLM, no programáis el modelo desde cero: **contratáis o accedéis a un servicio** que expone el modelo mediante una **API**. En el mercado coexisten muchos actores; cada uno ofrece modelos distintos, precios, límites y políticas de privacidad.

En la práctica del bootcamp y del TC os encontraréis sobre todo con:

| Tipo | Qué es | Ejemplo |
|------|--------|---------|
| **Proveedor de modelo + API** | Empresa que entrena o distribuye LLMs y los sirve en la nube | Google Gemini, OpenAI, Cohere |
| **Hub + inferencia** | Catálogo de modelos open source y APIs para ejecutarlos | Hugging Face |
| **Cloud hyperscaler** | IA integrada en la nube de AWS, Azure o Google Cloud | Amazon Bedrock, Azure OpenAI, Vertex AI |
| **Hosting de modelos abiertos** | Infraestructura que ejecuta Llama, Mistral, etc. | Groq, Together AI, Fireworks |

No hay un único “mejor” proveedor: depende de **coste**, **latencia**, **calidad en vuestro dominio**, **cumplimiento** (datos en UE, empresa, etc.) y de si ya usáis un ecosistema (Google, Microsoft, AWS).

### Ejemplo mínimo — una pregunta, una respuesta

Usamos **`ClientV2`** y el endpoint **`chat`**, el patrón actual del SDK de Cohere.

Parámetros clave:

- `model`: identificador del LLM en la plataforma Cohere
- `messages`: lista de turnos con roles `system`, `user`, `assistant`
- `temperature`: controla aleatoriedad (0.2 es razonable para onboarding, como en el benchmark del TC)

````python
import cohere

if not COHERE_API_KEY:
    raise ValueError("Falta COHERE_API_KEY. Ejecuta la celda anterior e introduce la clave.")

MODELO_COHERE = "command-r-plus-08-2024"  # alternativa ligera: command-r-08-2024
TEMPERATURE = 0.2

client = cohere.ClientV2(api_key=COHERE_API_KEY)

response = client.chat(
    model=MODELO_COHERE,
    messages=[
        {
            "role": "user",
            "content": "En una frase: ¿qué es un asistente de onboarding para empleados nuevos?",
        }
    ],
    temperature=TEMPERATURE,
)

texto = response.message.content[0].text
print(texto)
````

### Ejemplo contextual — rol de sistema + dominio Bridge SA

Patrón similar al que usaríais en `prompts.py` del Team Challenge: **system prompt fijo** + mensaje del usuario delimitado.

````python
SYSTEM_PROMPT = """
Eres el Employee Onboarding Assistant de Bridge SA.
Solo ayudas a empleados nuevos con accesos, políticas y primeros días.
No inventes políticas. Responde en español, breve.
""".strip()

pregunta_empleado = "¿A qué canales de Slack debo unirme el primer día?"

response = client.chat(
    model=MODELO_COHERE,
    messages=[
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": pregunta_empleado},
    ],
    temperature=TEMPERATURE,
)

print(response.message.content[0].text)
````

### Ejemplo con historial (varios turnos)

Ejemplo de chat con turnos `user` / `assistant`

````python
messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {
        "role": "user",
        "content": "Soy Laura, dev junior. Hoy es mi día 1 de onboarding.",
    },
]

r1 = client.chat(model=MODELO_COHERE, messages=messages, temperature=TEMPERATURE)
messages.append({"role": "assistant", "content": r1.message.content[0].text})

messages.append(
    {"role": "user", "content": "¿Y qué hago con GitHub antes de clonar repos?"}
)

r2 = client.chat(model=MODELO_COHERE, messages=messages, temperature=TEMPERATURE)
print(r2.message.content[0].text)
````

Streaming — respuesta token a token. Útil para demos en vivo. 

````python
stream = client.chat_stream(
    model=MODELO_COHERE,
    messages=[
        {"role": "user", "content": "Di hola en una línea, como asistente de onboarding."}
    ],
    temperature=TEMPERATURE,
)

for event in stream:
    if event.type == "content-delta":
        print(event.delta.message.content.text, end="", flush=True)

print()
````

### 🔑 Glosario exprés — antes de empezar con Hugging Face

| Término | En una frase |
|---|---|
| **Hub** | El catálogo de modelos de HF (huggingface.co/models) — no es un proveedor, es como el "GitHub de los modelos" |
| **Inference Providers** | El sistema de HF que reenvía tu petición a un proveedor externo (Groq, Together, Cerebras...) que ejecuta el modelo |
| **Router automático** | Lo que usa este notebook: `InferenceClient(api_key=...)` sin más — HF elige el proveedor por ti |
| **`provider="hf-inference"`** | Un backend concreto y limitado de HF (no el router). Catálogo pequeño, sin chat con Llama. En este bootcamp: **solo para embeddings** |
| **Modelo gated** | Modelo que exige aceptar una licencia en la web de HF antes de poder descargarlo |
| **Caché local** | Carpeta (`~/.cache/huggingface/hub/`) donde se guardan los pesos ya descargados, para no re-descargar cada vez |

> 💡 **Analogía:** *Inference Providers* es el aeropuerto entero (todas las puertas, todas las aerolíneas). `hf-inference` es una única terminal, pequeña y con pocos vuelos. Si no indicas `provider=`, el router te lleva a la puerta correcta automáticamente — eso es lo que queremos para chat.

## 2) Hugging Face — Chat LLM en la nube

### ¿Qué es Hugging Face Inference?

[Hugging Face](https://huggingface.co) es un hub de modelos **open source**. Para **chat en la nube** usamos `huggingface_hub.InferenceClient` con el **router de Inference Providers**: HF elige un proveedor serverless (Groq, Together, Cerebras, etc.) según el modelo.

```text
Tu notebook  →  InferenceClient (router)  →  proveedor serverless  →  texto generado
```

**No confundir con:**
- **Embeddings (Sprint 8):** `InferenceClient(provider="hf-inference")` + `feature_extraction()`.
- **Local (sección 3):** descargas el modelo y lo ejecutas en tu CPU/GPU — sin Inference API.

**Documentación útil:**

- [Guía Inference](https://huggingface.co/docs/huggingface_hub/guides/inference)

### Ejemplo mínimo — `chat_completion`

````python
from huggingface_hub import InferenceClient

if not HF_TOKEN:
    raise ValueError("Falta HF_TOKEN. Ejecuta la celda anterior e introduce el token.")

# Modelo del catálogo Inference Providers (https://huggingface.co/inference/models)
MODELO_HF = "openai/gpt-oss-20b"
TEMPERATURE_HF = 0.2
MAX_TOKENS = 250

# Router automático — NO uses provider="hf-inference" para chat LLM
hf_client = InferenceClient(api_key=HF_TOKEN)


def chat_hf(messages: list[dict], modelo: str = MODELO_HF):
    return hf_client.chat_completion(
        model=modelo,
        messages=messages,
        max_tokens=MAX_TOKENS,
        temperature=TEMPERATURE_HF,
    )


response = chat_hf(
    [
        {
            "role": "user",
            "content": "En una frase: ¿qué es un asistente de onboarding para empleados nuevos?",
        }
    ]
)

texto_hf = response.choices[0].message.content
print(f"Modelo: {MODELO_HF}")
print(texto_hf)
````

## 3) Hugging Face — Chat LLM en local (tu máquina)

### ¿Qué hacemos aquí?

Descargamos un modelo **open source** del Hub y lo ejecutamos con la librería **`transformers`** en tu **CPU o GPU**. No hace falta token de Inference Providers (salvo modelos *gated*).

```text
Tu notebook  →  transformers + PyTorch  →  modelo en RAM/VRAM  →  texto generado
```

| | HF nube (sección 2) | HF local (esta sección) |
|--|---------------------|-------------------------|
| **Librería** | `huggingface_hub` | `transformers` + `torch` |
| **Token** | Sí (`HF_TOKEN`) | No (modelo público) |
| **Descarga** | No (solo API) | Sí (~1 GB la primera vez) |
| **Velocidad** | Depende del proveedor | CPU: lenta; GPU: más rápida |
| **Uso en TC** | Válido para producción demo | Útil para prototipos sin API / sin cuota |

**Modelo de ejemplo:** `Qwen/Qwen2.5-0.5B-Instruct` (~0,5B parámetros, instruct, multilingüe). Alternativa: `TinyLlama/TinyLlama-1.1B-Chat-v1.0`.

> 💡 **¿Por qué no usar siempre local?** Un modelo grande (varios GB) puede tardar minutos en descargarse y ocupar mucha RAM al cargarse; en CPU la generación también es más lenta. Por eso el TC recomienda modelos pequeños (0,5B–2B parámetros) para prototipar.

> 💡 **Cuantización (si ves modelos "4-bit" o "GGUF"):** técnica para comprimir el modelo y que ocupe menos memoria, a cambio de algo de precisión. No la usamos en este TC, pero la veréis mencionada en fichas de modelos más grandes como `gemma-2-2b-it`.

### ¿Dónde encontrar otros modelos (local)?

1. **Buscador del Hub:** [huggingface.co/models](https://huggingface.co/models)  
   - Filtro **Task:** `Text Generation`  
   - Busca en el nombre: `Instruct`, `Chat` o `it` (instruction-tuned)  
   - Ordena por *Trending* o *Most downloads*

2. **Búsquedas que suelen funcionar en portátil (CPU):**  
   - [`Qwen2.5-0.5B-Instruct`](https://huggingface.co/models?search=qwen2.5+0.5b+instruct)

### Cargar modelo y función de chat local

- `EJECUTAR_LOCAL = False` por defecto (no descarga nada).
- Cambia a `True` para descargar y ejecutar en tu máquina.
- Usamos `apply_chat_template` para formatear mensajes `system` / `user` / `assistant`.

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

EJECUTAR_LOCAL = False  # True → descarga y ejecuta en tu máquina
MODELO_LOCAL = "Qwen/Qwen2.5-0.5B-Instruct"
MAX_NEW_TOKENS = 150
TEMPERATURE_LOCAL = 0.2

tokenizer_local = None
model_local = None
device_local = "cuda" if torch.cuda.is_available() else "cpu"


def chat_local(messages: list[dict], max_new_tokens: int = MAX_NEW_TOKENS) -> str:
    if model_local is None or tokenizer_local is None:
        raise RuntimeError("Modelo local no cargado. Ejecuta la celda con EJECUTAR_LOCAL = True.")

    prompt = tokenizer_local.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )
    inputs = tokenizer_local(prompt, return_tensors="pt").to(device_local)

    with torch.no_grad():
        output_ids = model_local.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=TEMPERATURE_LOCAL,
            pad_token_id=tokenizer_local.eos_token_id,
        )

    new_tokens = output_ids[0][inputs["input_ids"].shape[-1] :]
    return tokenizer_local.decode(new_tokens, skip_special_tokens=True).strip()


if EJECUTAR_LOCAL:
    print(f"Cargando {MODELO_LOCAL} en {device_local} (primera vez: descarga ~1 GB)...")
    tokenizer_local = AutoTokenizer.from_pretrained(MODELO_LOCAL)
    model_local = AutoModelForCausalLM.from_pretrained(
        MODELO_LOCAL,
        torch_dtype=torch.float32 if device_local == "cpu" else torch.float16,
    ).to(device_local)
    model_local.eval()
    print("Modelo local listo.")
else:
    print("HF local omitido (EJECUTAR_LOCAL = False). Cambia a True para probar.")
````

### Ejemplo mínimo — una pregunta

````python
if EJECUTAR_LOCAL:
    respuesta_local = chat_local(
        [
            {
                "role": "user",
                "content": "En una frase: ¿qué es un asistente de onboarding para empleados nuevos?",
            }
        ]
    )
    print(f"Modelo: {MODELO_LOCAL} ({device_local})")
    print(respuesta_local)
else:
    print("Activa EJECUTAR_LOCAL en la celda anterior.")
````
