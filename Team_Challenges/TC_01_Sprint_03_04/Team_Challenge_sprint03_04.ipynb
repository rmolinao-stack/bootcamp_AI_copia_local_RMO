## Parte 1: Data analytics (Pandas)

### 1. Ingesta de datos

- Los CSV están en **`Team_Challenges/TC_01_Sprint_03_04/data/`**.
- Cargar con **Pandas** los cuatro ficheros: `movies.csv`, `ratings.csv`, `tags.csv`, `links.csv`.
- Comprobar para cada `DataFrame`: `shape`, columnas, `dtypes`, `head` y conteo de nulos.

```python
# --- Apartado 1: Ingesta de datos ---

import pandas as pd
import pathlib

def procesar_csv(nombre_fichero: str):
    df = pd.read_csv(pathlib.Path(f"./data/{nombre_fichero}"))
    print(f"********* Datos para {nombre_fichero} ****************")
    print(f"El fichero tiene {df.shape[0]} filas y {df.shape[1]} columnas")
    print(f"El fichero tiene las siguientes columnas: {list(df.columns)}")
    print(f"Los tipos son los siguientes:\n{df.dtypes}")
    print("A continuación se muestras los cinco primeros registros:")
    display(df.head())
    print("A continuación se muestras los nulos por columnas:")
    display(df.isnull().sum())
    print("A continuación mostramos el dataFrame:")
    display(df)
    return df
    

df_movies = procesar_csv("movies.csv")
df_ratings = procesar_csv("ratings.csv")
df_tags = procesar_csv("tags.csv")
df_links = procesar_csv("links.csv")
```

### 2. Columna `year` desde el título

- En el dataset de películas. el año de estreno suele aparecer **entre paréntesis al final** de `title`, p. ej. `Batman (1989)`.
- Implementad una función (`year_from_title`) que devuelva un entero de cuatro cifras o valor ausente (`NaN` / `<NA>`) si el título no sigue ese patrón.
- Añadid la columna **`year`** a `movies` como numérico (`pd.to_numeric(..., errors="coerce")`).
- Editar el campo `title` para que no contenga el año de estreno.
- Contad cuántas películas **no** tienen año reconocible y mostrad **una pequeña muestra** de sus títulos (casos límite).

```python
# --- Apartado 2: columna `year` ---

def year_from_title(titulo: str) -> int:
    try:
        # Por python, cualquier cosa que no sea un entero perfecto
        # positivo o negativo dará error.
        # Hago strip para quitar espacios al final y cosas raras.
        year = int(titulo.strip()[-5:-1])
        if year < 0: raise ValueError
        return year
    except ValueError:
        return pd.NA
    
df_movies["year"] = df_movies["title"].apply(year_from_title)

# RAUL: 
# Con la llamda anterior estaría todo cubierto
# Si me obligan a usar pd.to_numeric habría que hacer lo siguiente
# pero no le encuentro sentido. Además visualmente saca decimales.
# df_movies["year"] = round(pd.to_numeric(df_movies["year"], errors="coerce"), 0)

df_no_year = df_movies[df_movies["year"].isnull()]
print(f"El número de películas sin año son: {len(df_no_year)}")

# RAUL: 
# Esto se puede hacer con df_no_year["title"], pero ya no
# es un dataframe y al lanzar display se ve raro. 
display(df_no_year.loc[:,["title"]])

```

### 2. Unificación de datos (*merge* / *join*)

- Entender las **claves** entre tablas (p. ej. `movieId` une `movies`, `ratings`, `tags` y `links`).
- Construir un esquema unificado: al menos una tabla **película–usuario–rating** (ratings enriquecida con título y géneros) y otra con **película–tags** si procede.
- Usar `merge` (u operaciones equivalentes) con criterio claro: tipo de unión (*inner* / *left*), duplicados generados y cómo se resuelven.
- Imprimir las 10 primeras filas de las tablas resultantes.
- Dejar documentado qué filas se pierden o se multiplican al unir y **por qué** es aceptable en vuestro caso de uso.

```python
# --- Apartado 3: Merge / join  ---

# RAUL: 
# En general el master será la película y querremos que siempre salga la película aunque haya nulos. 
df_mov_x_rat = pd.merge(df_movies, df_ratings, on="movieId", how="left")
df_mov_x_tag = pd.merge(df_movies, df_tags, on="movieId", how="left")
display(df_mov_x_rat)
display(df_mov_x_tag )
# RAUL - Juego de pruebas:
# Añadir una línea con movieID con 0 y verificar que sale con el resto nulos.
```

### 4. Agregaciones y segmentación

- Usar `groupby` (por usuario, por película o por género) con al menos una agregación multi-columna (`agg`).
- Comparar dos segmentos (p. ej. usuarios con muchas valoraciones vs. pocos; o por **década** usando la columna `year` del apartado 3) con tablas resumen.

```python
# --- Apartado 4: Agregaciones ---

# RAUL: Devuelve por película la nota media, la máxima y la mínima.
df_notas = df_mov_x_rat.groupby(by=["movieId", "title"])["rating"].agg(["mean", "max", "min"]).round({"mean": 2})
#display(df_notas)

# RAUL: Vamos a contar cuantas películas son estan aprobadas o suspendidas. 

df_aprobadas = df_notas[df_notas["mean"] >= 2.5]
df_aprobadas["Nota"] = "Aprobada"

df_suspendidas = df_notas[df_notas["mean"] < 2.5]
df_suspendidas["Nota"] = "Suspendida"

display(((pd.concat([df_aprobadas, df_suspendidas])).groupby("Nota").count())[["mean"]].rename(columns={"mean": "cantidad"}))

```
